```python
from warnings import filterwarnings
filterwarnings("ignore")
import pandas as pd
A=pd.read_csv("D:/python/project/training_set.csv")
A.head(2)
```

# Missing Data treatment

```python
def replacer(df):
    import pandas as pd
    Q = pd.DataFrame(df.isna().sum(),columns=["CT"])
    W = Q[Q.CT > 0].index
    for i in W:
        if(df[i].dtypes == "object"):
            x = df[i].mode()[0]
            df[i] = df[i].fillna(x)
        else:
            x = df[i].mean()
            df[i] = df[i].fillna(x)
            
```

```python
replacer(A)
```

```python
A.isna().sum()
```

# Drop Unnecessary columns

```python
for i in A.columns:
    print(i,len(A[i].unique()))
```

```python
B=A.drop(labels=("Id"),axis=1)
```

```python
Y = B[["SalePrice"]]
X = B.drop(labels=["SalePrice"],axis=1)
```

# Preprocessing

```python
def data_prep(X):
    
    cat = []
    con = []
    for i in X.columns:
        if(X[i].dtypes == "object"):
            cat.append(i)
        else:
            con.append(i)
            
            
    import pandas as pd
    X1 = pd.get_dummies(X[cat])
    from sklearn.preprocessing import StandardScaler
    ss = StandardScaler()
    X2 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
    X3 = X2.join(X1)
    return X3
```

```python
Xnew=data_prep(X)
```

# Divide Data in training and testing set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

# Create NN

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
nn = Sequential()
nn.add(Dense(200,input_dim=(Xnew.shape[1])))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(200))
nn.add(Dropout(0.2))
nn.add(Dense(1,activation="relu"))
```

```python
nn.compile(loss="mean_squared_error",metrics=["accuracy"])
nn.fit(xtrain,ytrain,validation_split=0.2,batch_size=100,epochs=1000)
```
